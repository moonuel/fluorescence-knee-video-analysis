Import file 

```python
import cv2
import numpy as np
import utils 

# Read video
fn = "data\\Annex.avi"
cap = cv2.VideoCapture(fn)

# Get video properties
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fourcc = int(cap.get(cv2.CAP_PROP_FOURCC))

# Copy all frames to a nicer object
video = np.empty((n_frames,h,w,3), dtype=np.uint8)
f_is_read = True
frame_num = 0
while cap.isOpened() and frame_num < n_frames:
    
    # Read frames
    f_is_read, frame = cap.read()

    # Write frames to array
    video[frame_num] = frame 

    frame_num += 1

# Close video cap
cap.release()
cv2.destroyAllWindows()
```

```python
# Show frames

for frame in video:

    # Show frame
    cv2.imshow("frame", frame)

    # Optionally, exit video by pressing 'q'
    if cv2.waitKey(100) == ord('q'):
        break

cv2.destroyAllWindows()
```

Apply centroid stabilization

```python
# Show frames

video_ctrd = []
for frame in video:

    # Get centered frame
    frame_ctrd, _ = utils.centroid_stabilization(frame, 100)

    # Store centered frame
    video_ctrd.append(frame_ctrd)

    # Optionally, exit video by pressing 'q'
    if cv2.waitKey(10) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
# Show frames

for frame in video_ctrd:

    # Show frame
    cv2.imshow("frame", frame)

    # Optionally, exit video by pressing 'q'
    if cv2.waitKey(10) == ord('q'):
        break

cv2.destroyAllWindows()
```

Write to file

```python
out_fn = "output\\Annex_centered.avi"
fourcc = cv2.VideoWriter_fourcc(*'XVID')
output_writer = cv2.VideoWriter(out_fn, fourcc, fps, (w,h))

for frame in video_ctrd:
    output_writer.write(frame)

output_writer.release()
```
